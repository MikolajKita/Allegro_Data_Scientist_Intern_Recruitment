```python
orders_delivered = orders[['order_id', 'customer_id', 'order_purchase_timestamp']][(orders['order_status']=='delivered')]
products_with_category = products[['product_category_name', 'product_id']].dropna()
orders_delivered = orders_delivered.merge(order_reviews[['review_score', 'order_id']], on='order_id', how = 'left')
orders_delivered = orders_delivered.merge(order_items[['order_id', 'product_id', 'price']], on='order_id', how = 'left')
orders_delivered = orders_delivered.merge(customer[['customer_id', 'customer_city', 'customer_state']], on='customer_id', how = 'left')
products_with_ratings = orders_delivered.merge(products_with_category, on='product_id', how = 'inner')
```

```python
orders_by_product = products_with_ratings.groupby('product_id').size().sort_values(ascending=False)
orders_average_price = products_with_ratings[['product_id', 'price']].groupby("product_id").agg("mean")
orders_by_product = orders_by_product.to_frame().reset_index()
orders_by_product.rename(columns={0: "number_of_purchases"}, inplace = True)
products_with_ratings = products_with_ratings.merge(orders_by_product, on='product_id', how = 'inner')
products_with_ratings.fillna(0)
orders_by_product = orders_by_product.merge(orders_average_price, on = 'product_id', how ='inner')
orders_by_product["revenue_per_product"] = (orders_by_product['price']*orders_by_product['number_of_purchases'])
orders_by_product.sort_values(by = 'revenue_per_product', inplace = True, ascending = False)
orders_by_product.reset_index(inplace=True)
```

```python
time_cutoff = '2017-07-27'
products_with_ratings.drop_duplicates(inplace=True)
products_with_ratings.sort_values(by='order_purchase_timestamp', inplace = True)
product_groups = products_with_ratings.groupby('product_category_name')['product_id'].count()
products_with_ratings = products_with_ratings[(products_with_ratings['order_purchase_timestamp'] > time_cutoff)]
products_with_ratings
```

```python
customer_categories = products_with_ratings[products_with_ratings.groupby(['customer_id'])['product_category_name'].transform('nunique') > 1]
customer_categories.drop_duplicates()
print(customer_categories.info())
categories_with_customers = customer_categories[['customer_id', 'product_category_name', 'product_id', 'review_score', 'customer_city', 'customer_state']]
categories_with_customers = categories_with_customers.sort_values(by='customer_id')
```

```python
def find_similar_people(category_1, category_2, category_3):
    customers_with_categories_1 = categories_with_customers['customer_id'][(categories_with_customers['product_category_name'] == category_1)]
    customers_with_categories_1.drop_duplicates(inplace=True)
    customers_with_categories_2 =categories_with_customers['customer_id'][(categories_with_customers['product_category_name'] == category_2)]
    customers_with_categories_2.drop_duplicates(inplace=True)
    customers_with_categories_3 =categories_with_customers['customer_id'][(categories_with_customers['product_category_name'] == category_3)]
    customers_with_categories_3.drop_duplicates(inplace=True)
    customers_with_categories_1 = customers_with_categories_1.to_frame().reset_index()
    customers_with_categories_2 = customers_with_categories_2.to_frame().reset_index()
    customers_with_categories_3 = customers_with_categories_3.to_frame().reset_index()
    category_1_with_category_2 = customers_with_categories_1.merge(customers_with_categories_2['customer_id'], on='customer_id')    
    category_1_with_category_3 = customers_with_categories_1.merge(customers_with_categories_3['customer_id'], on ='customer_id')
    category_2_with_category_3 = customers_with_categories_2.merge(customers_with_categories_3['customer_id'], on ='customer_id')
    all_categories_combined = category_1_with_category_2.merge(category_1_with_category_3['customer_id'], on='customer_id', how ='outer')
    all_categories_combined = all_categories_combined.merge(category_2_with_category_3['customer_id'], on='customer_id', how ='outer')
    all_categories_combined.drop_duplicates(inplace=True)
    return all_categories_combined
```

```python
def user_info(dataset, orders_by_product_with_revenue, number_to_display = 10, category_1=0, category_2=0, category_3=0, city=0, state=0):
    
    if category_1 == 0 and category_2 == 0 and category_3 == 0 and city == 0 and state == 0:
            
            return orders_by_product_with_revenue['product_id'].head(10)
    similar_people_id = find_similar_people(category_1, category_2, category_3)
    similar_people = similar_people_id.merge(dataset[['product_id', 'customer_id', 'product_category_name', 'review_score']], on='customer_id', how = 'inner')
    missing_amount_of_items = number_to_display - similar_people.shape[0]
    review_mtx = similar_people.pivot_table(values='review_score', index='customer_id', columns='product_id')
    review_mtx = review_mtx.mean().sort_values(ascending=False)
    if missing_amount_of_items <= 0:
        review_mtx = review_mtx.to_frame().reset_index()
        return review_mtx['product_id'].head(number_to_display)
    else:
        dataset = dataset[(((dataset['product_category_name'] == category_1) | (dataset['product_category_name'] == category_2) | 
                       (dataset['product_category_name'] == category_3)) & (dataset['customer_city']== city) & (dataset['customer_state'] == state))]
        dataset = dataset[['product_id', 'customer_id','review_score', 'product_category_name']]
        orders_grouped_by_product = dataset.groupby('product_id').size().sort_values(ascending=False)
        orders_grouped_by_product = orders_grouped_by_product.to_frame().reset_index()
        orders_grouped_by_product.rename(columns={0: "number_of_purchases"}, inplace = True)
        orders_grouped_by_product = orders_grouped_by_product[(orders_grouped_by_product['number_of_purchases'] > 3 * orders_grouped_by_product['number_of_purchases'].mean())]
        dataset = dataset.merge(orders_grouped_by_product, on='product_id', how = 'inner')
        review_mtx_same_location_one_category = dataset.pivot_table(values='review_score', index='customer_id', columns='product_id')
        review_mtx_same_location_one_category = review_mtx_same_location_one_category.mean().sort_values(ascending=False)
        dataset.drop_duplicates(inplace=True)
        review_mtx = review_mtx.to_frame().reset_index()
        review_mtx = review_mtx.rename(columns={0: "review_average_score"})
        review_mtx_same_location_one_category = review_mtx_same_location_one_category.to_frame().reset_index()
        review_mtx_same_location_one_category = review_mtx_same_location_one_category.rename(columns={0: "review_average_score"})
        if review_mtx.empty:
            return review_mtx_same_location_one_category['product_id'].head(number_to_display)        
        else:
            review_mtx_combined = pd.concat([review_mtx, review_mtx_same_location_one_category])
            review_mtx_combined = pd.DataFrame(review_mtx_combined)
            review_mtx_combined = review_mtx_combined[(review_mtx_combined['review_average_score']>=4)]
            review_mtx_combined.reset_index(inplace=True)
            return review_mtx_combined['product_id'].head(number_to_display)
            
```

```python
display_number = 10
product_matrix = user_info(products_with_ratings, orders_by_product, display_number, 'cama_mesa_banho', 'papelaria', 'fashion_calcados', 'sao paulo', 'SP')
product_matrix
```
